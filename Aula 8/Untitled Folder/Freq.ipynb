```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy import signal as sf
```

```python
Fs = 8000
```

```python
#Plotando a resposta em frequencia do exercicio 4 letra a
```

```python
#Usando freqz para obter a resposta em frequencia
num = [3, -3.6];
den = [1, -1.4, 0.45];
[w, H] = sf.freqz(num, den, Fs)
plt.plot(w*Fs/(2*np.pi), 20*np.log10(abs(H)))
plt.title('Resposta em frequencia A')
plt.grid
```
