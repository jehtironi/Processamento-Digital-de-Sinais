```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal as sf
```

```python
#Equação de Hamming

#w[i] = 0.54 - 0.46*cos(2*pi*i/M )

#Equação de Blackman

#wB[i]= 0.42 - 0.5*cos(2*pi*i/M ) + 0.08 * cos(4*pi*i/M )
```

```python
#Declarando as variaveis
```

```python
#tamanho do filtro
M = 100
w = np.zeros(100)
wB = np.zeros(100)
Fs = 8000
```

```python
#Laço de repetição para calcular Hamming
```

```python
for i in range(len(w)):
    w[i] = 0.54 - 0.46*np.cos(2*np.pi*i/M )
```

```python
#Laço de repetição para calcular Blackman
```

```python
for i in range(len(w)):
    w
    B[i]= 0.42 - 0.5 * np.cos(2*np.pi*i/M ) + 0.08 * np.cos(4*np.pi*i/M )
```

```python
#Plotando o grafico de saida
```

```python
# plota
plt.title(u"Gráfico Função Hamming")
plt.xlabel("amostra")
plt.ylabel("Amplitude")
plt.plot(w)
plt.show()
```

```python
# plota
plt.title(u"Gráfico Função Blackman")
plt.xlabel("amostra")
plt.ylabel("Amplitude")
plt.plot(w, color = 'pink')
plt.show()
```

```python
#Sobreposição de gráficos
```

```python
plt.title('Sobreposição de Gráficos')
plt.ylabel("Amplitude")
plt.grid(1)
plt.plot(w)
plt.plot(wB, color = 'pink') 
plt.tight_layout()
```

```python
#Plotando a resposta em frequencia do Hamming
```

```python
[p, w] = sf.freqz(w, 1, Fs)
plt.plot(p*Fs/(2*np.pi), 20*np.log10(abs(w)), label = 'Hamming')
plt.title('Resposta em frequencia Hamming')
```

```python
#Plotando a resposta em frequencia do Blackman
```

```python
[q, wB] = sf.freqz(wB, 1, Fs)
plt.plot(q*Fs/(2*np.pi), 20*np.log10(abs(wB)), label = 'Blackman')
plt.title('Resposta em frequencia Blackman')
```

```python
#Sobreposição dos graficos para verificar os resultados
#Atraves da plotação desse gráfico é evidente que o filtro de Blackman tem uma melhor atenuação
```

```python
plt.title('Resposta em frequencia Hamming X Blackman')
plt.plot(p*Fs/(2*np.pi), 20*np.log10(abs(w)))
plt.plot(q*Fs/(2*np.pi), 20*np.log10(abs(wB)), color = 'red')
plt.tight_layout()
```
