```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal as sf
```

```python
BAND-PASS WINDOWED-SINC FILTER
```

```python
#Iniciando as listas com todas as posições = 0
A = np.zeros(800)
B = np.zeros(800)
H = np.zeros(800)
```

```python
Fs = 8000 #Frequencia
FC = 0.196
tam = 100 #Tamanho do filtro
```

```python
for i in range(800):
    if(i-(tam/2) == 0):
        A[i] = 2* np.pi * FC
    if(i - tam/2) != 0:
        A[i] = np.sin(2 * np.pi * FC * (i-tam/2)) / (i - tam/2)
    A[i] = A[i] * (0.42-0.5)*np.cos(2*np.pi*(i/tam)) + 0.08*(np.cos(4 * np.pi*(i/tam)))
```

```python
SUM = 0
#Soma todos os coeficientes
for i in range(800):
    SUM = SUM + A[i] 

#Divide os coefiecientes pela soma de todos eles
for i in range(800):
    H[i] = A[i] / SUM
```

```python
FC = 0.204
for i in range(800):
    if(i-(tam/2) == 0):
        B[i] = 2* np.pi * FC
    if(i - tam/2) != 0:
        B[i] = np.sin(2 * np.pi * FC * (i-tam/2)) / (i - tam/2)
    B[i] = B[i] * (0.42-0.5)*np.cos(2*np.pi*(i/tam)) + 0.08*(np.cos(4 * np.pi*(i/tam)))
```

```python
SUM = 0
#Soma todos os coeficientes
for i in range(800):
    SUM = SUM + B[i] 

#Divide os coefiecientes pela soma de todos eles
for i in range(800):
    B[i] = B[i] / SUM
    
for i in range(800):
    B[i] = -B[i]

    B[400] = B[400+1]
    
for i in range(800):
    H[i] = A[i] + B[i]

for i in range(800):
    H[i] = -H[i]
    
H[400] = H[400+1]
        
```

```python
#Plotando o gráfico
plt.subplot(2, 1, 1)
plt.title("H[]")
plt.plot(H)
```

```python
[w, H] = sf.freqz(H, 1,Fs)
plt.plot(w*Fs/(2*np.pi), 20*np.log10(abs(H)))
plt.title('Resposta em frequencia')
plt.grid
```
