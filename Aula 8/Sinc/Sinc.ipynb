```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal as sf
```

```python
#função sinc é dada por h[i] = (sin(2*pi*Fc*i)/(i*pi))
```

```python
#declarando as variaveis
```

```python
Fs = 8000
Fc = 400

#cria uma lista
i = np.arange(-0.1, 0.1, 1/Fs)
```

```python
#Laço de repetição para calcular o h[i]
```

```python
h = []
h = np.zeros(len(i))
for j in range(0, len(i)):
    h[j] = np.sin(2*np.pi*fc*i[j])/(i[j]*np.pi)
```

```python
#Plotando o grafico da função sinc
```

```python
# plota
plt.title(u"Gráfico Função Sinc")
plt.xlabel("amostra")
plt.ylabel("Amplitude")
plt.plot(i, h)
plt.show()
```
