```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Declaração dos vetores
#Degrau Unitario
#Impulso unitario
```

```python
h = [0.125, 0.25, 0.375, 0.5, 0.625, 0.75, 0.875, 1]


#Impulso unitario
x = np.zeros(50)
x[0] = 1

#Degrau Unitario
y = np.ones(50)

#vetor
a = [1, 0.5, 0.25, 0.125]
```

```python
#Atraves da biblioteca de convolução realiza o calculo
```

```python
z = np.convolve(x, h)
w = np.convolve(y, h)
b = np.convolve(a, h)
```

```python
#plota o gráfico do sinal de entrada e de saida do impulso unitario
```

```python
plt.figure(figsize=(20, 13))
#Plotando o gráfico
plt.subplot(2, 2, 1)
plt.title("Sinal de Entrada Impulso")
plt.stem(x)

#Plotando o gráfico
plt.subplot(2, 2, 2)
plt.title("Sinal de Saida Impulso")
plt.stem(z)
```

```python
#plota o gráfico do sinal de entrada e de saida do degrau unitario
```

```python
plt.figure(figsize=(20, 13))
#Plotando o gráfico
plt.subplot(2, 2, 1)
plt.title("Sinal de Entrada Degrau")
plt.stem(y)

#Plotando o gráfico
plt.subplot(2, 2, 2)
plt.title("Sinal de Saida Degrau")
plt.stem(w)
```

```python
#Vetor
```

```python
plt.figure(figsize=(20, 13))
#Plotando o gráfico
plt.subplot(2, 2, 1)
plt.title("Sinal de Entrada Vetor")
plt.stem(a)

#Plotando o gráfico
plt.subplot(2, 2, 2)
plt.title("Sinal de Saida Vetor")
plt.stem(b)
```

```python
#Salvando os arquivos em um novo arquivo .pcm
```

```python
#Salvando o arquivo de saída
with open("ImpulsoUnitarioC.pcm", "wb") as novo:
    np.array(z, dtype=np.int16).tofile(novo)
novo.close()
    
    
#Salvando o arquivo de saída
with open("DegrauUnitarioC.pcm", "wb") as novo:
    np.array(w, dtype=np.int16).tofile(novo)
novo.close()

#Salvando o arquivo de saída
with open("VetorC.pcm", "wb") as novo:
    np.array(b, dtype=np.int16).tofile(novo)
novo.close()
```
