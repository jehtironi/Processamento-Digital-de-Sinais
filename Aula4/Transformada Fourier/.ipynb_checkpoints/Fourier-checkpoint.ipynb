```python
#Importando as bibliotecas
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Declarando o vetor w
#O mesmo vai de -pi até pi e o valor vai aumentando de pi/100 em pi/100
```

```python
w = np.arange(-np.pi, np.pi, np.pi/100)
a = 0.5
```

```python
#Declarando as funções utilizadas
```

```python
Num = 1
Den = 1 - a*np.exp(-1j*w)
X = Num/Den #x = numerador/denominador
```

```python
#Calculando o Modulo de X
```

```python
Mod_X = np.abs(X)
```

```python
#Calculando a Fase de X
```

```python
Fase_X = np.angle(X)
```

```python
#Plotando o módulo
```

```python
plt.subplot(2,1,1)
plt.plot(Mod_X,)
plt.title('Magnitude de X[n]')
```

```python
#Plotando a fase 
```

```python
#Plota a fase
plt.subplot(2,1,2)
plt.plot(Fase_X)
plt.title('Fase de X[n]')
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = 0.5
w = np.arange(-1*np.pi, 1*np.pi, np.pi/100)

num = 1
den = 1 - a * np.exp(-1j*w) # 1-a^(-j*w) 
x = num/den

mod_x = np.absolute(x)
fase_x = np.angle(den)

plt.figure("Transformada de Fourier",figsize=(15,8))
plt.subplot(211)
plt.plot(mod_x)
plt.title('Modulo')
plt.grid(True)

plt.subplot(212)
plt.plot(fase_x)
plt.title('Fase')
plt.grid(True)


plt.tight_layout()
plt.savefig("tfd.png", format="png")
plt.show()
```
