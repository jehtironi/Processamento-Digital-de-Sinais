```python
import numpy as np
import matplotlib.pyplot as plt

def impulsoUnitario():
    #Gerando o impulso
    x = np.arange(-5,5,1)
    impulso = np.zeros(10)
    impulso[5] = 1
       
    #Plotando o gráfico
    plt.subplot(2, 3, 1)
    plt.title('Impulso Unitario')
    plt.xlabel("Tempo[s]")
    plt.ylabel("Amplitude")
    plt.stem(x, impulso)
    
        
    #Salvando o arquivo de saída
    with open("impulsoUnitario.pcm", "wb") as novo:
        for data in (impulso):
            novo.write(data)
        novo.close()
    
    
def degrauUnitario():
    #Gerando o degrau
    
    n = np.arange(-5, 5,1)

    degrau = np.zeros(10)
    
    for i in range(0, 10):
        if(i < 5):
            degrau[i] = 0
        else:
            degrau[i] = 1
    
    #Plotando o gráfico
    plt.subplot(2, 3, 2)
    plt.title('Degrau Unitario')
    plt.xlabel("Tempo[s]")
    plt.ylabel("Amplitude")
    plt.stem(n, degrau)
    
    #Salvando o arquivo de saída
    with open("degrauUnitario.pcm", "wb") as novo:
        for data in (degrau):
            novo.write(data)
        novo.close()
    
    
    
def sequenciaSinusoidal():
    f0 = 100
    fs = 8000
    t = 1
       
    
    x = np.arange(-30, 30, 1)
    
    sinusoidal = t * np.cos(2*np.pi*x*(f0/fs))
    
    
    
    #Plotando o gráfico
    plt.subplot(2, 3, 3)
    plt.title('Sequencia Sinusoidal')
    plt.xlabel("Tempo[s]")
    plt.ylabel("Amplitude")
    plt.stem(x, sinusoidal) 
    
    #Salvando o arquivo de saída
    with open("Sinusoidal.pcm", "wb") as novo:
        for data in (sinusoidal):
            novo.write(data)
        novo.close()
    
    
    

def sequenciaExponencial():
    n = np.arange(0, 10, 0.2)
    
    exponencial = 1*(0.5**n)
    
    #Plotando o gráfico
    plt.subplot(2, 3, 4)
    plt.title('Sequencia Exponencial a = 0.5')
    plt.xlabel("Tempo[s]")
    plt.ylabel("Amplitude")
    plt.stem(n, exponencial) 
    
    #Salvando o arquivo de saída
    with open("exponencial1.pcm", "wb") as novo:
        for data in (exponencial):
            novo.write(data)
        novo.close()
      
    exponencial = 1*(-0.5**n)
    
    #Plotando o gráfico
    plt.subplot(2, 3, 5)
    plt.title('Sequencia Exponencial a = - 0.5')
    plt.xlabel("Tempo[s]")
    plt.ylabel("Amplitude")
    plt.stem(n, exponencial) 
    
    #Salvando o arquivo de saída
    with open("exponencial2.pcm", "wb") as novo:
        for data in (exponencial):
            novo.write(data)
        novo.close()
    
    exponencial = 1*(2**n)
    
    #Plotando o gráfico
    plt.subplot(2, 3, 6)
    plt.title('Sequencia Exponencial a = 2')
    plt.xlabel("Tempo[s]")
    plt.ylabel("Amplitude")
    plt.stem(n, exponencial) 
    
    #Salvando o arquivo de saída
    with open("exponencial3.pcm", "wb") as novo:
        for data in (exponencial):
            novo.write(data)
        novo.close()
    
plt.figure(figsize=(20, 13))
impulsoUnitario()
degrauUnitario()
sequenciaSinusoidal()
sequenciaExponencial()
    
```
