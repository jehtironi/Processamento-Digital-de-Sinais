```python
#Declarando variaveis
#Solicita ao usuario o tamanho
```

```python
M = 100
K = int(input("K = "))
Fc = 400
h = np.zeros(M)
Fs = 8000
```

```python
#Laço de repetição para calcular os valores de h[i]
#Dentro do laço de repetição possui uma condicionalaonde quando o i for igual a metade de M o valor de H[i] sera dado por 
#h[i] = (2*np.pi*fc)*K, isso é feito para evitar uma divisão por zero
```

```python
for i in range(len(h)):
     if(i == M/2): 
          h[i] = (2*np.pi*fc)*K
     else: 
          h[i] = K * (numpy.sin(2*numpy.pi*fc*(i-(M/2)))/(i-(M/2))) * ((0.42 - 0.5 * numpy.cos((2*numpy.pi)*i/M)) + 0.08*numpy.cos((4*numpy.pi*i)/M))
```

```python
#Plotando os graficos de saida
```

```python
# plota
plt.title(u"Gráfico")
plt.xlabel("amostra")
plt.ylabel("Amplitude")
plt.plot(h)
plt.show()
```
