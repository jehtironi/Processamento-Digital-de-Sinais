```python
import numpy as np
import matplotlib.pyplot as plt

h = [1, 0.5, 0.25, 0.125]

#Impulso unitario
x = np.zeros(2000)
x[0] = 1

#Degrau Unitario
y = np.ones(2000)

z = np.convolve(x, h)
w = np.convolve(y, h)

plt.figure(figsize=(20, 13))
#Plotando o gráfico
plt.subplot(2, 2, 1)
plt.title("Sinal de Entrada Impulso")
plt.stem(x)

#Plotando o gráfico
plt.subplot(2, 2, 2)
plt.title("Sinal de Saida Impulso")
plt.stem(z)

plt.figure(figsize=(20, 13))
#Plotando o gráfico
plt.subplot(2, 2, 1)
plt.title("Sinal de Entrada Degrau")
plt.stem(y)

#Plotando o gráfico
plt.subplot(2, 2, 2)
plt.title("Sinal de Saida Degrau")
plt.stem(w)

#Salvando o arquivo de saída
with open("impulsoFinal.pcm", "wb") as novo:
    for data in z:
        novo.write(data)
    novo.close()
    
#Salvando o arquivo de saída
with open("DegrauC.pcm", "wb") as novo:
    for data in w:
        novo.write(data)
    novo.close()

```
