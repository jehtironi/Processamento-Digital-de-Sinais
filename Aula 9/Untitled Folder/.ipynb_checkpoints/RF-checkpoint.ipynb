```python
import numpy as np
from scipy import signal as sf
import matplotlib.pyplot as plt

#Declarando as variaveis
Fc1 = 400
Fc2 = 800
Fs = 8000
transicao = 200

Fc1 = Fc1/Fs
Fc2 = Fc2/Fs
M = int(4/(transicao/Fs))

#zerando as listas
h1 = np.zeros(M)
h2 = np.zeros(M)
hRF = np.zeros(M)

# Calculando o coeficiente passa baixa
for i in range(M):
    if((i-M/2) == 0):
        h1[i] = 2*np.pi*Fc1

    elif((i-M/2) != 0):
        h1[i] = np.sin(2*np.pi*Fc1*(i-(M/2))) / (i-(M/2))
    h1[i] = h1[i] * (0.54 - 0.46 * np.cos(2 * np.pi * i/M))

sum = 0
for i in range(M):
    sum = sum + h1[i]

for i in range(M):
    h1[i] = h1[i] / sum

# # Calculando o coeficiente passa Alta
for i in range(M):
    if((i-M/2) == 0):
        h2[i] = 2*np.pi*Fc2
    elif((i-M/2) != 0):
        h2[i] = np.sin(2*np.pi*Fc2*(i-(M/2))) / (i-(M/2))

    h2[i] = h2[i] * (0.54 - 0.46 * np.cos(2 * np.pi * i/M))

sum = 0
for i in range(M):
    sum = sum + h2[i]

for i in range(M):
    h2[i] = h2[i] / sum

for i in range(M):
    h2[i] = -(h2[i])

i = int(M / 2)
h2[i] = h2[i] + 1

# Calculando os coeficientes do rejeita faixa
for i in range(M):
    hRF[i] = h1[i] + h2[i]


with open("CoefRF.dat", "w") as f:
    for s in hRF:
        f.write(str(s) +",\n")


# Plotando os grficos
[w1, h1] = sf.freqz(h1, worN = Fs, fs = Fs)
[w2, h2] = sf.freqz(h2, worN = Fs, fs = Fs)

plt.subplot(2, 1, 1)
plt.title("Resposta de frequncia Windowed-sinc")
plt.plot(w1, abs(h1))
plt.plot(w2, abs(h2))
plt.legend()
plt.xlabel("Freq")
plt.ylabel("Amplitude")

plt.subplot(2, 1, 2)
plt.title("Resposta de frequncia (dB)")
plt.plot(w1, 20 * np.log10(abs(h1)), label="freqz1")
plt.plot(w2, 20 * np.log10(abs(h2)), label="freqz2")
plt.legend()
plt.xlabel("Freq")
plt.ylabel("Amplitude (dB)")

        
plt.tight_layout()
plt.grid()
plt.show()
```
