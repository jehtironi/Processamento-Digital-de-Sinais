```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Função para realizar o Impulso Unitario
#A lista x possui um tamanho de 200, e com a função np.zeros ela sera completada com zeros.
#Como é um impulso unitario, no x[0] a mesma recebe o valor numero 1;
#Em seguida é plotado o gráfico e salvado o arquivo resultando.
```

```python
def impulsoUnitario():
    #Gerando o impulso
    x = np.zeros(200)
    x[0] = 1

    #Plotando o gráfico
    plt.subplot(2, 3, 1)
    plt.title('Impulso Unitario')
    plt.xlabel("Tempo[s]")
    plt.ylabel("Amplitude")
    plt.stem(x)


    #Salvando o arquivo de saída
    with open("impulsoUnitario.pcm", "wb") as novo:
        np.array(x, dtype=np.int16).tofile(novo)
    novo.close()
```

```python
#Função para realizar o Degrau Unitario
#Atraves da função np.arange é criado uma lista que vai do -100 até o 100, aumentando de 1 em 1;
#Em seguida temos um for para alterar os valores, sendo que abaixo de 0 o valor sera 0 e acima e igual a 0 o degrau recebe 1
#Em seguida é plotado o gráfico e salvado o arquivo resultando.
```

```python
def degrauUnitario():
    #Gerando o degrau
    n = np.arange(-100, 100, 1)

    degrau = np.zeros(200)

    for i in range(-100, 100):
        if(i < 0):
            degrau[i] = 1
        else:
            degrau[i] = 0

    #Plotando o gráfico
    plt.subplot(2, 3, 2)
    plt.title('Degrau Unitario')
    plt.xlabel("Tempo[s]")
    plt.ylabel("Amplitude")
    plt.stem(n, degrau)

    #Salvando o arquivo de saída
    with open("DegrauUnitario.pcm", "wb") as novo:
        np.array(degrau, dtype=np.int16).tofile(novo)
    novo.close()
```

```python
#Função para realizar Sequencia Sinusoidal
#Utilizado a função do np.cos para calcular a seuqencia
#Em seguida é plotado o gráfico e salvado o arquivo resultando.
```

```python
def sequenciaSinusoidal():
    f0 = 100
    fs = 8000
    t = 1
       
    
    x = np.arange(-100, 100, 1)
    
    sinusoidal = t * np.cos(2*np.pi*x*(f0/fs))
    
    
    
    #Plotando o gráfico
    plt.subplot(2, 3, 3)
    plt.title('Sequencia Sinusoidal')
    plt.xlabel("Tempo[s]")
    plt.ylabel("Amplitude")
    plt.stem(x, sinusoidal) 
    
    #Salvando o arquivo de saída
    with open("SequenciaSinusoidal.pcm", "wb") as novo:
        np.array(sinusoidal, dtype=np.int16).tofile(novo)
    novo.close()
```

```python
#Função para realizar a Sequencia Exponencial
#em primeiro momento é criado um array que inicia no 0 e aumenta de 0.2 em 0.2 até o valor final que é 10;
#Em seguida é realizado o calculo com o valor da base de 0.5, plotado o gráfico e salvado o arquivo;
#Em seguida é realizado o calculo com o valor da base de -0.5, plotado o gráfico e salvado o arquivo;
#Em seguida é realizado o calculo com o valor da base de 2, plotado o gráfico e salvado o arquivo;
```

```python
def sequenciaExponencial():
    n = np.arange(0, 10, 0.2)
    
    exponencial = 1*(0.5**n)
    
    #Plotando o gráfico
    plt.subplot(2, 3, 4)
    plt.title('Sequencia Exponencial a = 0.5')
    plt.xlabel("Tempo[s]")
    plt.ylabel("Amplitude")
    plt.stem(n, exponencial) 
    
    #Salvando o arquivo de saída
    with open("Exponencial1.pcm", "wb") as novo:
        np.array(exponencial, dtype=np.int16).tofile(novo)
    novo.close()
      
    exponencial = 1*(-0.5**n)
    
    #Plotando o gráfico
    plt.subplot(2, 3, 5)
    plt.title('Sequencia Exponencial a = - 0.5')
    plt.xlabel("Tempo[s]")
    plt.ylabel("Amplitude")
    plt.stem(n, exponencial) 
    
    #Salvando o arquivo de saída
    with open("Exponencial2.pcm", "wb") as novo:
        np.array(exponencial, dtype=np.int16).tofile(novo)
    novo.close()
    
    exponencial = 1*(2**n)
    
    #Plotando o gráfico
    plt.subplot(2, 3, 6)
    plt.title('Sequencia Exponencial a = 2')
    plt.xlabel("Tempo[s]")
    plt.ylabel("Amplitude")
    plt.stem(n, exponencial) 
    
    #Salvando o arquivo de saída
    with open("Exponencial3.pcm", "wb") as novo:
        np.array(exponencial, dtype=np.int16).tofile(novo)
    novo.close()
    
```

```python
#Chamada das funções para mostrar os resultados
```

```python
plt.figure(figsize=(20, 13))
impulsoUnitario()
degrauUnitario()
sequenciaSinusoidal()
sequenciaExponencial()
```
